# Load the datasets from csv

```python
users = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/users.csv')
users.head()
```

```python
sessions =pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/sessions.csv')
sessions.head()
```

```python
products = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/products.csv')
products.head()
```

```python
transactions =pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/transactions.csv')
transactions.head()
```

# <a id='7'>12. Join users to transactions, keeping all rows from transactions and only matching rows fromusers (left join)<a/>

```python
pd.merge(transactions , users , how = 'left')
```

# [13. Which transactions have a UserID not in users?](#0)

```python
transactions[~transactions['UserID'].isin(users['UserID'])]
```

# [14. Join users to transactions, keeping only rows from transactions and users that match via UserID (inner join) ](#0)

```python
pd.merge(transactions , users , how = 'inner')
```

# [15. Join users to transactions, displaying all matching rows AND all non-matching rows (full outer join)](#0)

```python
pd.merge(transactions , users , how = 'outer')
```

# [16. Determine which sessions occurred on the same day each user registered](#0)

```python
# First i merged both tables before doing the analysis to see how the tables look together before moving forward .
pd.merge(users , sessions , how = 'inner')
```

# Answer : These are the sessions that occured on the same day each user registered.

```python
pd.merge(left =users, right=sessions,how='inner',left_on=['UserID','Registered'],right_on = ['UserID','SessionDate'])
```

# [17.Build a dataset with every possible (UserID, ProductID) pair (cross join](#0)

```python
new_df = users.assign(value = 1).merge(products.assign(value = 1)).drop('value' , 1)
new_df.head(10)
```

```python
# Final Output
new_df2 = new_df[['UserID' , 'ProductID']]
new_df2
```

# [18. Determine how much quantity of each product was purchased by each user ](#0)

```python
user_product_qty = pd.merge(new_df2 , transactions , how = 'left' , on = ['UserID' , 'ProductID'])

user_qty = user_product_qty.groupby(['UserID','ProductID'])['Quantity'].sum()

user_qty.reset_index().fillna(0)
```

# [19. For each user, get each possible pair of pair transactions (TransactionID1, TransacationID2)](#0)

```python
pd.merge(transactions , transactions , how ='outer' , on = 'UserID')
```

# [20. Join each user to his/her first occuring transaction in the transactions table](#0)

```python
data = pd.merge(users , transactions.groupby('UserID').first().reset_index(), how = 'left' , on = 'UserID')
data.head()
```

# [21. Test to see if we can drop columns](#0)


```python
my_columns = list(data.columns) 
my_columns
```

```python
list(data.dropna(thresh=int(data.shape[0] * .9), axis=1).columns)
```

```python
missingInfo = list(data.columns[data.isnull().any()])
missingInfo
```

### [Missing information](#0)

```python
for col in missingInfo:
    missingNumber = data[data[col].isnull() == True].shape[0]
    print('Missing Number for Col {}: {}'.format(col , missingNumber))
```

### [Percentage of missing info](#0)

```python
for col in missingInfo:
    percentMissing = data[data[col].isnull() == True].shape[0] / data.shape[0]
    print('Col Percent Missing {}: {}'.format(col, percentMissing))
```
