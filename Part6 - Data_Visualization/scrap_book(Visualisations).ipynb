```python
fig , ((ax1 , ax2) , (ax3 , ax4)) = plt.subplots(nrows=2,
                                                ncols=2,
                                                figsize=(10,5))

# plot each axis
ax1.plot(x, x/2);
ax1.set(title="Line Plot",
       ylabel="Temperature")


ax2.scatter(np.random.random(10) , np.random.random(10));
ax2.set(title="Random",
       ylabel="Blalallalalblalab")


ax3.bar(choco.keys() , choco.values());
ax3.set(title="Chocalate bars",
       ylabel="Price")


ax4.hist(np.random.randn(1000));
ax4.set(title="Random Normal Dist",
       ylabel="Count")

plt.show()
```

# Distribution of male and femal with heart diseases

```python
gender_val = ['Male','Female']
gender = data['sex'].value_counts()

gender = gender.astype(float)
fig = plt.figure(figsize=(8,4))
ax = fig.add_axes((0,0,0.5,1))
ax.set_title('Gender Distribution' , bbox={'facecolor':'white','pad':5})
labels = 'Male','Female'
colors = ['blue' , 'red']
explode = (0.1 , 0)
plt.pie(gender ,explode =explode ,labels=labels ,colors =colors ,autopct = '%1.1f%%' ,shadow = True,startangle = 60)
plt.legend(labels = gender_val)
plt.axis('equal')
plt.show()
```

# Let's build on the above

```python
fig , ax = plt.subplots(figsize=(10,6))
# plot the data
scatter = ax.scatter(x=over_50["age"],
            y = over_50["chol"],
            c = over_50["target"]);
           
ax.set(title="Heart Diseases and Cholestrol levels",
      xlabel="Age",
      ylabel="Cholestrol") ; 

ax.legend(*scatter.legend_elements() , title="Target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
         linestyle="--",
         color="red");
            
```

```python
# Subplots for chol , age , thalach
```

```python
fig , (ax0 , ax2) = plt.subplots(nrows=2,
                                  ncols=1,
                                  figsize=(10,10))
```

```python
fig , (ax0 , ax1) = plt.subplots(nrows=2,
                                  ncols=1,
                                  figsize=(10,10))

# plot the data
scatter = ax0.scatter(x=over_50["age"],
            y = over_50["chol"],
            c = over_50["target"])
           
ax0.set(title="Heart Diseases and Cholestrol levels",
      xlabel="Age",
      ylabel="Cholestrol") ; 

ax0.legend(*scatter.legend_elements() , title="Target")

# Add a horizontal line
ax0.axhline(over_50["chol"].mean(),
         linestyle="--",
         color="red");

# plot the data
scatter = ax1.scatter(x=over_50["age"],
            y = over_50["thalach"],
            c = over_50["target"])
           
ax1.set(title="Heart Diseases and Max Heart Rate",
      xlabel="Age",
      ylabel="Max Heart Rate") ; 

ax1.legend(*scatter.legend_elements() , title="Target")

# Add a horizontal line
ax1.axhline(over_50["thalach"].mean(),
         linestyle="--",
         color="blue");

fig.suptitle("Heart Disease Analysis" , fontsize=16 , fontweight ="bold")
plt.show()
```

```python
fig , (ax0 , ax1) = plt.subplots(nrows=2,
                                  ncols=1,
                                  figsize=(10,10))

# plot the data
scatter = ax0.scatter(x=over_50["age"],
            y = over_50["chol"],
            c = over_50["target"])
           
ax0.set(title="Heart Diseases and Cholestrol levels",
      xlabel="Age",
      ylabel="Cholestrol") ; 

ax0.legend(*scatter.legend_elements() , title="Target")

# Add a horizontal line
ax0.axhline(over_50["chol"].mean(),
         linestyle="--",
         color="red");

# plot the data
scatter = ax1.scatter(x=over_50["age"],
            y = over_50["thalach"],
            c = over_50["target"],
                     cmap="winter")  # viridis ,
           
ax1.set(title="Heart Diseases and Max Heart Rate",
      xlabel="Age",
      ylabel="Max Heart Rate") ; 

ax1.legend(*scatter.legend_elements() , title="Target")

# Add a horizontal line
ax1.axhline(over_50["thalach"].mean(),
         linestyle="--",
         color="red");

fig.suptitle("Heart Disease Analysis" , fontsize=16 , fontweight ="bold")
plt.show()
```

```python
import os
print(os.listdir("../data"))
```

```python
df1 = pd.read_csv("../data/Chicago_Crimes_2005_to_2007.csv" ,error_bad_lines=False)
df2 = pd.read_csv("../data/Chicago_Crimes_2008_to_2011.csv" ,error_bad_lines=False)
df3 = pd.read_csv("../data/Chicago_Crimes_2012_to_2017.csv" ,error_bad_lines=False)
```

```python
data2 = pd.concat([df1, df2, df3], ignore_index=False , axis=0)
data2.head()
```

```python
# ID Case Number Date Block IUCR Primary Type Description Location Description Arrest Domestic Beat District Ward Community Area FBI Code X Coordinate Y Coordinate Year Updated On Latitude Longitude Location
data2.drop(['Unnamed: 0', 'Case Number', 'Case Number', 'IUCR', 'X Coordinate', 'Y Coordinate','Updated On','Year', 'FBI Code', 'Beat','Ward','Community Area', 'Location', 'District', 'Latitude' , 'Longitude'], inplace=True, axis=1)
```

```python
data2.Date
```

```python
data2.Date = pd.to_datetime(data2.Date , format="%m/%d/%Y %I:%M:%S %p")
```

```python
data2.index = pd.DatetimeIndex(data2.Date)
```
