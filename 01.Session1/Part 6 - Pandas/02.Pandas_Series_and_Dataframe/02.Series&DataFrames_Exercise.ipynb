# This is done by dealing with whitespace. It might not be obvious at first glance. But pay attention to the exits column

+ Trying to call the exit column will result in an error. 

```python
def clean_column(column_name):
    cleaned = column_name.strip().title()
    return cleaned
```

```python
# This is a list comprehension. It applies your clean function to every item in the list.
# We then reassign that to df.columns
# You shouldn't have to change anything here.
# Your function above should work appropriately here.
data.columns = [clean_column(col) for col in data.columns]
```

```python
# Check the result. Exits column donesnt look like this anymore 'Exits     '.    
data.columns
```
