# Distrution of male and female with heart diseases

```python
gen_val = ["Male","Female"]
gender = data["sex"].value_counts()

gender = gender.astype(float)
fix = plt.figure(figsize=(10,5))
ax = fig.add_axes((0,0,0.5,1))
ax.set_title("Gender" , bbox = {"facecolor":"white" , "pad":5})
labels = "Male" , "Female"
colors = ["purple" , "red"]
explode = (0.1, 0)
plt.pie(gender, explode=explode, labels=labels , colors=colors , 
        autopct = "%1.1f%%" , shadow= True,startangle=60)
plt.legend(labels = gen_val)
plt.axis("equal")
plt.show()
```

```python
# Lets build on the above
```

```python
fix , ax = plt.subplots(figsize=(10,6))
# plot the data
scatter = ax.scatter(
            x=over_50["age"],
            y=over_50["chol"],
            c=over_50["target"]);
           
ax.set(title="Heart Disease of over 50's vs Cholestrol levels",
      xlabel="Age",
      ylabel="Cholesterol");
    
ax.legend(*scatter.legend_elements() , title="Target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
         linestyle="--",
         color = "red");

plt.show()
```

# Suplots for chol , age , thalach

```python
fig , (ax0 , ax2) = plt.subplots(nrows=2,
                                ncols=1,
                                figsize=(10,6))
```

```python
fig , (ax0 , ax2) = plt.subplots(nrows=2,
                                ncols=1,
                                figsize=(10,10))
# plot the data
scatter = ax0.scatter(
            x=over_50["age"],
            y=over_50["chol"],
            c=over_50["target"]);
           
ax0.set(title="Heart Disease of over 50's vs Cholestrol levels",
      xlabel="Age",
      ylabel="Cholesterol");
    
ax0.legend(*scatter.legend_elements() , title="Target")

# Add a horizontal line
ax0.axhline(over_50["chol"].mean(),
         linestyle="--",
         color = "red");

# plot the data for thalach
scatter = ax2.scatter(
            x=over_50["age"],
            y=over_50["thalach"],
            c=over_50["target"],
            cmap="winter");
           
ax2.set(title="Heart Disease of over 50's vs Thalach levels",
      xlabel="Age",
      ylabel="Thalach");
    
ax2.legend(*scatter.legend_elements() , title="Target")

# Add a horizontal line
ax2.axhline(over_50["thalach"].mean(),
         linestyle="--",
        color="blue");

fig.suptitle("Heart Disease Analysis" , fontsize=17 , fontweight="bold")
plt.show()
```
