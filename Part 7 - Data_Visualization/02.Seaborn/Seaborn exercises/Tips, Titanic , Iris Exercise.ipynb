We will use **tips**, **iris** and another very famous dataset **titanic** for these exercises. These datasets are included in the seaborn's built-in dataset library.<br>
Let's do some exploratory data analysis using seaborn and present useful plots for data visualization!<br>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

**Do the following tasks:** 
* set style to '`white`'

```python
sns.set_style('white')
```

**load titanic dataset into '`titanic`', tips into '`tips`' and iris into '`iris`'**

```python
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
print ("\nTitanic data\n",'-'*50, sep='')
display(titanic = sns.load_dataset('titanic'))
display(titanic.head())


print ("\nLoading the Tips data\n",'-'*80, sep='')
display(tips = sns.load_dataset('tips'))
display(tips.head())


print ("\nLoading the Iris data\n",'-'*52, sep='')
iris = sns.load_dataset('iris')
iris.head()
```

# Tips data - Create a plot to see how much tip is left for each day for both gender

```python
sns.boxplot(x="day", y="tip", data=tips, hue = 'sex');
```

```python
# Create a jointplot for total bill and tip, set kind to "hex"
```

```python
sns.jointplot(x='total_bill',y='tip', data=tips, kind='hex');
```

# Titanic Data - Use a histogram to create a distribution for fare column

```python
sns.distplot(titanic['fare'],bins=20,kde=False,color='blue');
```

# Create a boxplot for class and age for survivors

```python
sns.boxplot(x='class',y='age',data=titanic,hue='alive',palette='coolwarm');
```

# Create a boxplot for class and age

```python
sns.boxplot(x='class',y='age',data=titanic,palette='rainbow');
```

**Recreate the above using swarm plot using titanic dataset, use factorplot** 

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
sns.factorplot(x='class',y='age',data=titanic, col="sex",kind="swarm",palette='rainbow');
```

**Can you plot how many male and female were on titanic? Separate them based on their class.**

```python
sns.countplot(x='sex',data=titanic,hue="class");
```

# Improving the above plot

```python
ax = sns.catplot(x="class" , y="survived", hue="sex" , data=titanic, height=6, 
            kind="bar" , palette="Set1")
ax.despine(left=True)
ax.set_ylabels("survival probability");
```

# Create a jointplot for both fare and age

```python
sns.jointplot(x='fare',y='age',data=titanic);
```

```python
g = sns.FacetGrid(data=titanic,col='sex', hue='alive')
g.map(plt.hist,'age',alpha=0.5)
g.add_legend();
```

# Create a heatmap for the correlation between variables

```python
sns.heatmap(titanic.corr(),cmap='YlGnBu',annot=True)
plt.title('titanic.correlation()')
plt.show()
```

# Iris Data - Pairplot

```python
g = sns.PairGrid(iris , hue="species")
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
g.add_legend();
```
